```python
from collections import defaultdict



# Словарь для хранения результата
data_dict = defaultdict(list)


filename = "mapping_inter.txt"
with open(filename, "r") as file:
    next(file)
    
    # Построчное чтение
    for line in file:
        # Убираем лишние пробелы и переносы строк
        line = line.strip()
        
        # Разделяем строку на Ozon ID и Артикул
        if line:  # Проверяем, что строка не пустая
            ozon_id, sku = line.split('\t')
            ozon_id = int(ozon_id)  # Преобразуем Ozon ID в число
            
            data_dict[ozon_id].append(sku)

filename = "mapping_at.txt"
with open(filename, "r") as file:
    next(file)
    
    # Построчное чтение
    for line in file:
        # Убираем лишние пробелы и переносы строк
        line = line.strip()

        if '\t' not in line:
            no_match.append(line)
            continue
        
        # Разделяем строку на Ozon ID и Артикул
        if line:  # Проверяем, что строка не пустая
            ozon_id, sku = line.split('\t')
            ozon_id = int(ozon_id)  # Преобразуем Ozon ID в число
            
            data_dict[ozon_id].append(sku)
```

```python
 import requests
 import pandas as pd
 import numpy as np
 from time import sleep


 def fetch_stocks(seller_legal: str, spreadsheet_id: str, sku_sheet_id: str, sku_sheet_name: str):
     """
     Получает данные об остатках через API Ozon и добавляет их в Google Sheets.

     :param seller_legal: Юридическое лицо.
     :param spreadsheet_id: Идентификатор Google Sheets для сохранения данных.
     :param sku_sheet_id: Идентификатор Google Sheets для получения списка SKU.
     :param sku_sheet_name: Название вкладки в Google Sheets для получения списка SKU.
     """

     log_prefix = f"Остатки. {seller_legal}. "
     err_prefix = f"\033[91mWARNING: {log_prefix}\033[0m"

     # Получаем API-ключ и client_id
     api_key = CONFIG["marketplace_keys"].get(seller_legal)
     client_id = CONFIG["ozon"]["client_ids"].get(seller_legal)

     if not api_key:
         raise Exception(f"API ключ для юридического лица '{seller_legal}' не найден в конфигурации.")
     if not client_id:
         raise Exception(f"Client ID для юридического лица '{seller_legal}' не найден в конфигурации.")

     headers = {
         "Client-Id": client_id,
         "Api-Key": api_key
     }


     if not skus_from_match:
         raise Exception(f"{err_prefix}Список SKU пуст. Проверьте данные в таблице.")

     # Выгружаем данные об остатках с использованием API
     offset = 0
     combined_data = []

     while True:
         payload = {
             "filter": {
                 "offer_ids": skus,
                 "warehouse_ids": []
             },
             "limit": 1000,
             "offset": offset
         }

         while True:
             response = requests.post("https://api-seller.ozon.ru/v1/analytics/manage/stocks", json=payload,
                                      headers=headers)
             #print(f"Статус: {response.status_code}, Ответ: {response.text[:500]}")  # Ограничиваем вывод текста
             if response.status_code == 429:
                 print(f"{err_prefix}Достигнут лимит запросов. Ожидание 60 секунд...")
                 sleep(60)
             elif response.status_code != 200:
                 raise Exception(f"{err_prefix}Ошибка получения данных об остатках: {response.status_code}, {response.text}")
             else:
                 break

         # Парсим данные из ответа
         data = response.json().get("items")
         next_offset = response.json().get("next")

         # Если данных нет, или `next` равен None, выходим из цикла
         if not data:
             #print("Данные закончились или отсутствуют.")
             break

         combined_data.extend(data)

         # Проверяем наличие следующей страницы
         if not next_offset:
             #print("Следующих страниц нет, завершение цикла.")
             break

         # Обновляем offset для следующего запроса
         offset += 1000

     if not combined_data:
         raise Exception(f"{err_prefix}Не удалось получить данные об остатках. Проверьте запросы и входные данные.")

     # Преобразование данных в DataFrame
     final_data = pd.DataFrame(combined_data)

     if final_data.empty:
         raise Exception(f"{err_prefix}Полученный DataFrame пуст. Проверьте входные данные и API-ответ.")

     # Преобразование NaN и недопустимых значений для JSON совместимости
     final_data = final_data.replace([np.inf, -np.inf], np.nan)

     # Разделяем столбцы по типу данных
     numeric_cols = final_data.select_dtypes(include=[np.number]).columns
     non_numeric_cols = final_data.select_dtypes(exclude=[np.number]).columns

     # Для числовых столбцов заменяем NaN на 0
     final_data[numeric_cols] = final_data[numeric_cols].fillna(0)

     # Для нечисловых столбцов заменяем NaN на ""
     final_data[non_numeric_cols] = final_data[non_numeric_cols].fillna("")

     # Дополнительно убеждаемся, что больше нет NaN значений
     assert not final_data.isnull().values.any(), f"{err_prefix}DataFrame содержит NaN значения после обработки."

     # Загружаем данные в Google Sheets без заголовков и с форматированием
     print(f"{log_prefix}Данные добавлены на вкладку StocksReport(ext).")
     return final_data
```
