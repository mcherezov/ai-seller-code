```python
import getpass

db_config = {
    "dbname": 'app_db',
    "user": 'user',
    "password": getpass.getpass(),
    "host": '93.183.73.107',
    "port": 57212
}
```

```python
pg_date_from = '2024-02-02'
pg_date_to = '2024-02-03'
query = "SELECT sku, date, cost_in_kopecks FROM cost WHERE date >= %s AND date <= %s"
```

```python
import psycopg2

result = []
try:
    # Подключение к базе данных
    with psycopg2.connect(**db_config) as conn:
        with conn.cursor() as cur:
            # Выполнение запроса
            cur.execute(query, (pg_date_from, pg_date_to))
            
            # Извлечение данных
            rows = cur.fetchall()
            for row in rows:
                result.append({'product_id': row[0], row[1].strftime('%d.%m.%Y'): row[2] / 100})
except psycopg2.Error as e:
    raise Exception(f"Ошибка при работе с базой данных: {e}")
```
